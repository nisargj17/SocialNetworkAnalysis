```python
"""
Task 3: Cleaning-up the network
        write code that reports the connected components of a graph."""


"""
The DFS algorithm works as follows:

1.Start by putting any one of the graph's vertices on top of a stack.
2.Take the top item of the stack and add it to the visited list.
3.Create a list of that vertex's adjacent nodes. Add the ones which aren't in the visited list to the top of stack.
4.Keep repeating steps 2 and 3 until the stack is empty"""

#DFS for to keep note of every node in graph
def dfs_task3(graph, node):
    seen = {node}
    stack = [node]
    result = []
    while stack:
        x = stack[-1]
        nxt = set(graph.get(x, [])) - seen
        if nxt:
            stack.extend(list(nxt))
            seen |= nxt
        else:
            result.append(x)
            stack.pop()
    return tuple(result)


#Adancency list to store edges keep track of visited nodes
def adjancency_list(edges):
    adjlist = {}
    for node in sorted(edges):
        if node[0] not in adjlist:
            adjlist[node[0]] = [node[1]]
        else:
            adjlist[node[0]].append(node[1])
    for node in sorted(edges):
        if node[1] not in adjlist:
            adjlist[node[1]] = []
    return adjlist


def cleaning(edges):
    visited = set()
    pathDictionary = {}
    adj_list = adjancency_list(edges)
    # vertices        
    test = []
    for key in adj_list:
        if key in visited: 
            print("Ignoring n {} beacuse we have visited already ".format(key))
            continue 
        results = dfs_task3(adj_list,key)
        for n in results: 
            visited.add(n)
        test.append(results)
    return test

#main
adj_list = adjancency_list(te)
results  = cleaning(te)
in_network = []

for i in results:
    in_network.append(len(i))
result1 = results[0]
result2 = results[1]
inside_network = list(set(result1))
outside_network = list(set(result2) - set(result1))
print("Outside of Network",outside_network)
print("Inside Network",inside_network)
```

```python
"""
Task 4: Closed subnetworks
        write code that computes the decomposition of the network into strongly connected components"""



"""
Following is Kosarajus DFS based simple algorithm that does two DFS traversals of graph:
1) Initialize all vertices as not visited.

2) Do a DFS traversal of graph starting from any arbitrary vertex v. If DFS traversal doesnt visit all vertices, then return false.

3) Reverse all arcs (or find transpose or reverse of graph)

4) Mark all vertices as not-visited in reversed graph.

5) Do a DFS traversal of reversed graph starting from same vertex v (Same as step 2). If DFS traversal doesnt visit all vertices, then return false. Otherwise return true."""

class Kosaraju (object):
    def __init__(self, te):
        #Convert edgelist to adjancency nodelist, with forward and reverse direction
        self.nodelist={}
        for i in range(len(te)):
            #if edgelist is given nodeid as int, then use append()
            if te[i][0] != te[i][1]:
                if (te[i][0] in self.nodelist) & (te[i][1] in self.nodelist) : 
                    self.nodelist[te[i][0]]['connected_node'].extend([te[i][1]])
                    self.nodelist[te[i][1]]['connected_node_reverse'].extend([te[i][0]])
                elif te[i][0] in self.nodelist:
                    self.nodelist[te[i][0]]['connected_node'].extend([te[i][1]])
                    self.nodelist[te[i][1]]={'connected_node_reverse':[te[i][0]],'connected_node':[]}
                elif te[i][1] in self.nodelist:
                    self.nodelist[te[i][1]]['connected_node_reverse'].extend([te[i][0]])
                    self.nodelist[te[i][0]]={'connected_node':[te[i][1]],'connected_node_reverse':[]}    
                else:
                    self.nodelist[te[i][1]]={'connected_node_reverse':[te[i][0]],'connected_node':[]} 
                    self.nodelist[te[i][0]]={'connected_node':[te[i][1]],'connected_node_reverse':[]}    
        #for all node, initialize the finishing time, leader, explored flag
        for key in self.nodelist:
            self.nodelist[key]['finishing_time']=0
            self.nodelist[key]['explored']=False
        #scc dictionary used to store all strongly connected components. Key as leader component    
        self.scc={}
        
    def get_scc_kosaraju(self):
        #Step1, First pass in reverse direction
        #populate the node loop stack in the order of first pass finishing time
        self.node_loop_stack=[]
        for node in self.nodelist:
            if self.nodelist[node]['explored']==False:
                self.DFS(node,True)
                
        #Step2, Second pass in forward direction
        #reset the explored flag to False, and initialize the leader as None
        for key in self.nodelist:
            self.nodelist[key]['leader']=None
            self.nodelist[key]['explored']=False
        #Loop through all the nodes in the order of reverse of first pass finishing time
        while self.node_loop_stack != []:
            node = self.node_loop_stack.pop()
            if self.nodelist[node]['explored']==False:
                self.leader=node
                self.scc[self.leader]={'component_node':[]}                
                self.DFS(node,False)        
                
    def DFS(self, node, reverse_flag):
        if reverse_flag==True:
            self.nodelist[node]['explored']=True
            for child_node in self.nodelist[node]['connected_node_reverse']:
                if self.nodelist[child_node]['explored']==False:
                    self.DFS(child_node, True)
            self.node_loop_stack.extend([node])

        else:
            self.nodelist[node]['explored']=True
            for child_node in self.nodelist[node]['connected_node']:
                if self.nodelist[child_node]['explored']==False:
                    self.DFS(child_node, False)  
            self.scc[self.leader]['component_node'].extend([node])    
```

```python
"""
Task 5: A recommendation engine
        Write code that takes as input two personsP and P and a number k, finds the number of length-k following chains that connect P to P ."""

class Recommendation:

    X = []
    Matrix  = []
    
    def __init__(SelfG, edges, m):
        SelfG.edges =  edges
        SelfG.m     = m
        
    def Adj_Matrix(SelfG):
        data = SelfG.edges
        nodes = np.unique(data) 
        noidfs = {n: i for i, n in enumerate(nodes)}  
        n = nodes.size 
        numdata = np.vectorize(noidfs.get)(data)  
        X = np.zeros((n, n))
        for tail, head in numdata:
            X[tail, head] = 1
        Recommendation.X = X 
        return Recommendation.X
    
    def matrix_multiplication(SelfG):
        Recommendation.Matrix = np.linalg.matrix_power(Recommendation.X, SelfG.m)
        return Recommendation.Matrix
    
    def length_k_chains(SelfG, i, j):
        p1 = i 
        p2 = j 
        return Recommendation.Matrix[p1][p2] 
```
